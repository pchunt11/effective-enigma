```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
url1 =requests.get("https://jang.com.pk/").text
```

```python
from datetime import datetime
def jang_text():
    now = datetime.now()
    date = now.strftime("%d_%m_%y")
    
    url1 =requests.get("https://jang.com.pk/").text
    
    soup = BeautifulSoup(url1, "lxml")
    html = soup.prettify()

    html_name = "Jang_html/DJ_" + date + ".html"
    with open(html_name, "w", encoding="utf-8") as file:
        file.write(html)

    
    dj_fp = soup.find_all('section', class_='main-story top-featured-news-main')
    dj_stories = dj_fp[0].find_all('li')
    
    list_titles = []
    list_links = []
    content =[]

    for i in np.arange(0,len(dj_stories)):
        title = dj_stories[i].get_text()
        list_titles.append(title)
        
        link = dj_stories[i].find('a')['href']
        list_links.append(link)
        
        article = requests.get(link)
        art_content = article.content
        soup_art = BeautifulSoup(art_content, 'lxml')
        body = soup_art.find_all('div', 'description-area')
        x = body[0].find_all('p')
          
        list_paragraphs = []
        for p in np.arange(0, len(x)):
            par = x[p].get_text()
            list_paragraphs.append(par)
            final_article = " ".join(list_paragraphs)
        
        content.append(final_article)
        
        txt_file = "Jang/DJ_" + date + ".txt"
        with open(txt_file, 'w', encoding="utf-8") as f:
            for item in content:
                f.write("%s\n" % item)
    
jang_text()
```
